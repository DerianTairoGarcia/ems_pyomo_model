```python
from math import acos, atan, cos, sin, sqrt	
# Function to calculate Parameters

PDa = {}
PDb = {}
PDc = {}
QDa = {}
QDb = {}
QDc = {}
# Transforms the demands in variate values in the time
List_NT = []
for i in N:
	for t in T:
		List_NT += [[i, t]]
for (i,t) in List_NT:
	PDa[(i,t)] = ' '
	PDb[(i,t)] = ' '
	PDc[(i,t)] = ' '
	QDa[(i,t)] = ' '
	QDa[(i,t)] = ' '
	QDa[(i,t)] = ' '

for (i,t) in List_NT:
	PDa[(i,t)] = fm_pa[i][t] * PDa_0[i] 
	PDb[(i,t)] = fm_pb[i][t] * PDb_0[i] 
	PDc[(i,t)] = fm_pc[i][t] * PDc_0[i]
	QDa[(i,t)] = fm_qa[i][t] * QDa_0[i]
	QDb[(i,t)] = fm_qb[i][t] * QDb_0[i]
	QDc[(i,t)] = fm_qc[i][t] * QDc_0[i]
PVa = {}
PVb = {}
PVc = {}
for (i,t) in List_NT:
	PVa[(i,t)] = ' '
	PVb[(i,t)] = ' '
	PVc[(i,t)] = ' '
# Transforms the PV generation in variate values in the time
for (i,t) in List_NT:
	PVa[(i,t)] = fpv[t] * PVa_0[i]
	PVb[(i,t)] = fpv[t] * PVb_0[i]
	PVc[(i,t)] = fpv[t] * PVc_0[i]
for (i,j) in L:
	Raa[(i,j)] = Raa[(i,j)]/1000
	Rbb[(i,j)] = Rbb[(i,j)]/1000
	Rcc[(i,j)] = Rcc[(i,j)]/1000
	Rab[(i,j)] = Rab[(i,j)]/1000
	Rac[(i,j)] = Rac[(i,j)]/1000
	Rbc[(i,j)] = Rbc[(i,j)]/1000
	Xaa[(i,j)] = Xaa[(i,j)]/1000
	Xbb[(i,j)] = Xbb[(i,j)]/1000
	Xcc[(i,j)] = Xcc[(i,j)]/1000
	Xab[(i,j)] = Xab[(i,j)]/1000
	Xac[(i,j)] = Xac[(i,j)]/1000
	Xbc[(i,j)] = Xbc[(i,j)]/1000
# Calculats the impedance magnitude and angle in the lines
Thaa = {}
Zaa =  {}
Thbb = {}
Zbb =  {}
Thcc = {}
Zcc =  {}
Thab = {}
Zab =  {}
Thac = {}
Zac =  {}
Thbc = {}
Zbc =  {}
for (i,j) in L:
	Zaa[(i,j)] = ' '
	Zbb[(i,j)] = ' '
	Zcc[(i,j)] = ' '
	Zab[(i,j)] = ' '
	Zac[(i,j)] = ' '
	Zbc[(i,j)] = ' '
	Thaa[(i,j)] = ' '
	Thbb[(i,j)] = ' '
	Thcc[(i,j)] = ' '
	Thab[(i,j)] = ' '
	Thac[(i,j)] = ' '
	Thbc[(i,j)] = ' '
for (i,j) in L:
	Thaa[(i,j)] = atan(Xaa[(i,j)] / Raa[(i,j)])
	Zaa[(i,j)] = sqrt(Raa[(i,j)]**2 + Xaa[(i,j)]**2)
	Thbb[(i,j)] = atan(Xbb[(i,j)] / Rbb[(i,j)])
	Zbb[(i,j)] = sqrt(Rbb[(i,j)]**2 + Xbb[(i,j)]**2)
	Thcc[(i,j)] = atan(Xcc[(i,j)] / Rcc[(i,j)])
	Zcc[(i,j)] = sqrt(Rcc[(i,j)]**2 + Xcc[(i,j)]**2)
	Thab[(i,j)] = 0
	Zab[(i,j)] = sqrt(Rab[(i,j)]**2 + Xab[(i,j)]**2)
	Thac[(i,j)] = 0
	Zac[(i,j)] = sqrt(Rac[(i,j)]**2 + Xac[(i,j)]**2)
	Thbc[(i,j)] = 0
	Zbc[(i,j)] = sqrt(Rbc[(i,j)]**2 + Xbc[(i,j)]**2)
# Defines the angles of the phases
Tha0 = {}
Thb0 = {}
Thc0 = {}
for i in N:
	Tha0[i] = 0
	Thb0[i] = -2.0944
	Thc0[i] = 2.0944
# Calculation of Transformed Impedance Components
Raa_p = {}
Xaa_p = {}
Rbb_p = {}
Xbb_p = {}
Rcc_p = {}
Xcc_p = {}
Rab_p = {}
Xab_p = {}
Rac_p = {}
Xac_p = {}
Rbc_p = {}
Xbc_p = {}
Rba_p = {}
Xba_p = {}
Rca_p = {}
Xca_p = {}
Rcb_p = {}
Xcb_p = {}

for (i,j) in L:
	Raa_p[(i,j)] = ' '
	Xaa_p[(i,j)] = ' '
	Rbb_p[(i,j)] = ' '
	Xbb_p[(i,j)] = ' '
	Rcc_p[(i,j)] = ' '
	Xcc_p[(i,j)] = ' '
	Rab_p[(i,j)] = ' '
	Xab_p[(i,j)] = ' '
	Rac_p[(i,j)] = ' '
	Xac_p[(i,j)] = ' '
	Rbc_p[(i,j)] = ' '
	Xbc_p[(i,j)] = ' '
	Rba_p[(i,j)] = ' '
	Xba_p[(i,j)] = ' '
	Rca_p[(i,j)] = ' '
	Xca_p[(i,j)] = ' '
	Rcb_p[(i,j)] = ' '
	Xcb_p[(i,j)] = ' '
for (i,j) in L:
	Raa_p[(i,j)] = Zaa[(i,j)] * cos(Thaa[(i,j)] + Tha0[i] - Tha0[i])
	Xaa_p[(i,j)] = Zaa[(i,j)] * sin(Thaa[(i,j)] + Tha0[i] - Tha0[i])
	Rbb_p[(i,j)] = Zbb[(i,j)] * cos(Thbb[(i,j)] + Thb0[i] - Thb0[i])
	Xbb_p[(i,j)] = Zbb[(i,j)] * sin(Thbb[(i,j)] + Thb0[i] - Thb0[i])
	Rcc_p[(i,j)] = Zcc[(i,j)] * cos(Thcc[(i,j)] + Thc0[i] - Thc0[i])
	Xcc_p[(i,j)] = Zcc[(i,j)] * sin(Thcc[(i,j)] + Thc0[i] - Thc0[i])
	Rab_p[(i,j)] = Zab[(i,j)] * cos(Thab[(i,j)] + Thb0[i] - Tha0[i])
	Xab_p[(i,j)] = Zab[(i,j)] * sin(Thab[(i,j)] + Thb0[i] - Tha0[i])
	Rac_p[(i,j)] = Zac[(i,j)] * cos(Thac[(i,j)] + Thc0[i] - Tha0[i])
	Xac_p[(i,j)] = Zac[(i,j)] * sin(Thac[(i,j)] + Thc0[i] - Tha0[i])
	Rbc_p[(i,j)] = Zbc[(i,j)] * cos(Thbc[(i,j)] + Thc0[i] - Thb0[i])
	Xbc_p[(i,j)] = Zbc[(i,j)] * sin(Thbc[(i,j)] + Thc0[i] - Thb0[i])
	Rba_p[(i,j)] = Zab[(i,j)] * cos(Thab[(i,j)] + Tha0[i] - Thb0[i])
	Xba_p[(i,j)] = Zab[(i,j)] * sin(Thab[(i,j)] + Tha0[i] - Thb0[i])
	Rca_p[(i,j)] = Zac[(i,j)] * cos(Thac[(i,j)] + Tha0[i] - Thc0[i])
	Xca_p[(i,j)] = Zac[(i,j)] * sin(Thac[(i,j)] + Tha0[i] - Thc0[i])
	Rcb_p[(i,j)] = Zbc[(i,j)] * cos(Thbc[(i,j)] + Thb0[i] - Thc0[i])
	Xcb_p[(i,j)] = Zbc[(i,j)] * sin(Thbc[(i,j)] + Thb0[i] - Thc0[i])		
```

```python

#ACTIVE POWER BALANCE WITHOUT CONTINGENCES

def active_power_balance_rule_a(model_cs, i,t):
    return (
        sum(model_cs.Pa[a, j, t]*df[(i,a,j)] for (a,j) in L) +
        model_cs.Ppcc_a[i, t] + sum(model_cs.PGa[a, t] for a in dict_nos_gd[i]) - PDa[(i,t)] == 0
    )

model_cs.active_power_balance_a = Constraint(List_NT, rule=active_power_balance_rule_a)

def active_power_balance_rule_b(model_cs, i,t):
    return (
        sum(model_cs.Pb[a, j, t]*df[(i,a,j)] for (a,j) in L) +
        model_cs.Ppcc_b[i, t] + sum(model_cs.PGb[a, t] for a in dict_nos_gd[i]) -
        PDb[(i,t)] == 0
    )

model_cs.active_power_balance_b = Constraint(List_NT, rule=active_power_balance_rule_b)

def active_power_balance_rule_c(model_cs, i,t):
    return (
        sum(model_cs.Pc[a, j, t]*df[(i,a,j)] for (a,j) in L) +
        model_cs.Ppcc_c[i, t] + sum(model_cs.PGc[a, t] for a in dict_nos_gd[i]) -
        PDc[(i,t)] == 0
    )

model_cs.active_power_balance_c = Constraint(List_NT, rule=active_power_balance_rule_c)

#REACTIVE POWER BALANCE

def reactive_power_balance_rule_a(model_cs, i,t):
    return (
        sum(model_cs.Qa[a, j, t]*df[(i,a,j)] for (a,j) in L) +
        model_cs.Qpcc_a[i, t] + sum(model_cs.QGa[a, t] for a in dict_nos_gd[i]) -
        QDa[(i,t)] == 0
    )

model_cs.reactive_power_balance_a = Constraint(List_NT, rule=reactive_power_balance_rule_a)

def reactive_power_balance_rule_b(model_cs, i,t):
    return (
        sum(model_cs.Qb[a, j, t]*df[(i,a,j)] for (a,j) in L) +
        model_cs.Qpcc_b[i, t] + sum(model_cs.QGb[a, t] for a in dict_nos_gd[i])-
        QDb[(i,t)] == 0
    )

model_cs.reactive_power_balance_b = Constraint(List_NT, rule=reactive_power_balance_rule_b)

def reactive_power_balance_rule_c(model_cs, i,t):
    return (
        sum(model_cs.Qc[a, j, t]*df[(i,a,j)] for (a,j) in L) +
        model_cs.Qpcc_c[i, t]+ sum(model_cs.QGc[a, t] for a in dict_nos_gd[i]) -
        QDc[(i,t)] == 0
    )

model_cs.reactive_power_balance_c = Constraint(List_NT, rule=reactive_power_balance_rule_c)


#Genset 

def genset_power_rule(model,i, t):
    return(model_cs.PGa[i, t] + model_cs.PGb[i, t] + model_cs.PGc[i, t] == model_cs.PG[i, t])
model_cs.genset_power_active = Constraint(List_GDT, rule=genset_power_rule)

def genset_power_reactive_rule(model,i, t):
    return(model_cs.QGa[i, t] + model_cs.QGb[i, t] + model_cs.QGc[i, t] == model_cs.QG[i, t])
model_cs.genset_power_reactive = Constraint(List_GDT, rule=genset_power_reactive_rule)

def genset_power_active_limits_rule_1(model,i, t):
    return(model_cs.PG[i, t] >= PG_min[i])
model_cs.genset_power_active_limits_1 = Constraint(List_GDT, rule = genset_power_active_limits_rule_1)

def genset_power_active_limits_rule_2(model,i, t):
    return(model_cs.PG[i, t] <= PG_max[i])
model_cs.genset_power_active_limits_2 = Constraint(List_GDT, rule = genset_power_active_limits_rule_2)

def genset_power_reactive_limits_rule_1(model,i, t):
    return(model_cs.QG[i, t] >= QG_min[i])
model_cs.genset_power_reactive_limits_1 = Constraint(List_GDT, rule = genset_power_reactive_limits_rule_1)

def genset_power_reactive_limits_rule_2(model,i, t):
    return(model_cs.QG[i, t] <= QG_max[i])
model_cs.genset_power_reactive_limits_2 = Constraint(List_GDT, rule = genset_power_reactive_limits_rule_2)

def genset_operation_1_rule(model,i, t):
    return(model_cs.PGa[i, t] == 0)
model_cs.genset_operation_1 = Constraint(List_GDT, rule = genset_operation_1_rule)

def genset_operation_2_rule(model,i, t):
    return(model_cs.PGb[i, t] == 0)
model_cs.genset_operation_2 = Constraint(List_GDT, rule = genset_operation_2_rule)

def genset_operation_3_rule(model,i, t):
    return(model_cs.PGc[i, t] == 0)
model_cs.genset_operation_3 = Constraint(List_GDT, rule = genset_operation_3_rule)

def genset_operation_4_rule(model,i, t):
    return(model_cs.QGa[i, t] == 0)
model_cs.genset_operation_4 = Constraint(List_GDT, rule = genset_operation_4_rule)

def genset_operation_5_rule(model,i, t):
    return(model_cs.QGb[i, t] == 0)
model_cs.genset_operation_5 = Constraint(List_GDT, rule = genset_operation_5_rule)

def genset_operation_6_rule(model,i, t):
    return(model_cs.QGc[i, t] == 0)
model_cs.genset_operation_6 = Constraint(List_GDT, rule = genset_operation_6_rule)

#FIX VARIABLES

def fix_voltage_a_rule(model_cs,i,t):
    return(model_cs.Va[i,t] == Vnom)
model_cs.fix_voltage_a = Constraint(List_NT, rule = fix_voltage_a_rule)

def fix_voltage_b_rule(model_cs,i,t):
    return(model_cs.Vb[i,t] == Vnom)
model_cs.fix_voltage_b = Constraint(List_NT, rule = fix_voltage_b_rule)

def fix_voltage_c_rule(model_cs,i,t):
    return(model_cs.Vc[i,t] == Vnom)
model_cs.fix_voltage_c = Constraint(List_NT, rule = fix_voltage_c_rule)

model_cs.fix_active_power = ConstraintList()
for i in Tb:
    for t in T:
        if Tb[i] != 1:
            model_cs.fix_active_power.add(expr = model_cs.Ppcc_a[i,t] == 0)
            model_cs.fix_active_power.add(expr = model_cs.Ppcc_b[i,t] == 0)
            model_cs.fix_active_power.add(expr = model_cs.Ppcc_c[i,t] == 0)
            
model_cs.fix_reactive_power = ConstraintList()
for i in Tb:
    for t in T:
        if Tb[i] != 1:
            model_cs.fix_reactive_power.add(expr = model_cs.Qpcc_a[i,t] == 0)
            model_cs.fix_reactive_power.add(expr = model_cs.Qpcc_b[i,t] == 0)
            model_cs.fix_reactive_power.add(expr = model_cs.Qpcc_c[i,t] == 0)
```

```python
#ACTIVE POWER BALANCE

def active_power_balance_rule_a_out(model_cs, i, t, o):
    return (
        sum(model_cs.Pa_out[a, j, t, o]* df[(i,a,j)] for (a,j) in L) +
        sum(model_cs.PGa_out[a, t, o] for a in dict_nos_gd[i]) +
        model_cs.Ppcc_a_out[i, t, o] - PDa[(i,t)] * model_cs.xd[i,t,o] == 0
    )

model_cs.active_power_balance_a_out = Constraint(List_NTO, rule=active_power_balance_rule_a_out)

def active_power_balance_rule_b_out(model_cs, i, t, o):
    return (
        sum(model_cs.Pb_out[a, j, t, o]* df[(i,a,j)] for (a,j) in L) +
        sum(model_cs.PGb_out[a, t, o] for a in dict_nos_gd[i]) +
        model_cs.Ppcc_b_out[i, t, o] -
        PDb[(i,t)] * model_cs.xd[i,t,o] == 0
    )

model_cs.active_power_balance_b_out = Constraint(List_NTO, rule=active_power_balance_rule_b_out)

def active_power_balance_rule_c_out(model_cs, i, t, o):
    return (
        sum(model_cs.Pc_out[a, j, t, o]* df[(i,a,j)] for (a,j) in L) +
        sum(model_cs.PGc_out[a, t, o] for a in dict_nos_gd[i]) +
        model_cs.Ppcc_c_out[i, t, o] -
        PDc[(i,t)] * model_cs.xd[i,t,o] == 0
    )

model_cs.active_power_balance_c_out = Constraint(List_NTO, rule=active_power_balance_rule_c_out)

#REACTIVE POWER BALANCE

def reactive_power_balance_rule_a_out(model_cs, i,t,o):
    return (
        sum(model_cs.Qa_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) +
        sum(model_cs.QGa_out[a, t, o] for a in dict_nos_gd[i]) +
        model_cs.Qpcc_a_out[i, t, o] -
        QDa[(i,t)] * model_cs.xd[i,t,o] == 0
    )

model_cs.reactive_power_balance_a_out = Constraint(List_NTO, rule=reactive_power_balance_rule_a_out)

def reactive_power_balance_rule_b_out(model_cs, i,t,o):
    return (
        sum(model_cs.Qb_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) +
        sum(model_cs.QGb_out[a, t, o] for a in dict_nos_gd[i]) +
        model_cs.Qpcc_b_out[i, t, o] -
        QDb[(i,t)] * model_cs.xd[i,t,o] == 0
    )

model_cs.reactive_power_balance_b_out = Constraint(List_NTO, rule=reactive_power_balance_rule_b_out)

def reactive_power_balance_rule_c_out(model_cs, i,t,o):
    return (
        sum(model_cs.Qc_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) +
        sum(model_cs.QGc_out[a, t, o] for a in dict_nos_gd[i]) +
        model_cs.Qpcc_c_out[i, t, o] -
        QDc[(i,t)] * model_cs.xd[i,t,o] == 0
    )

model_cs.reactive_power_balance_c_out = Constraint(List_NTO, rule=reactive_power_balance_rule_c_out)

#Genset Operation

def genset_power_rule_out(model,i, t, o):
    return(model_cs.PGa_out[i, t, o] + model_cs.PGb_out[i, t, o] + model_cs.PGc_out[i, t, o] == model_cs.PG_out[i, t, o])
model_cs.genset_power_active_out = Constraint(List_GDTO, rule=genset_power_rule_out)

def genset_power_reactive_rule_out(model,i, t, o):
    return(model_cs.QGa_out[i, t, o] + model_cs.QGb_out[i, t, o] + model_cs.QGc_out[i, t, o] == model_cs.QG_out[i, t, o])
model_cs.genset_power_reactive_out = Constraint(List_GDTO, rule=genset_power_reactive_rule_out)

def genset_power_active_limits_rule_out_1(model,i, t, o):
    return(model_cs.PG_out[i, t, o] >= PG_min[i])
model_cs.genset_power_active_limits_1_out = Constraint(List_GDTO, rule = genset_power_active_limits_rule_out_1)

def genset_power_active_limits_rule_out_2(model,i, t, o):
    return(model_cs.PG_out[i, t, o] <= PG_max[i])
model_cs.genset_power_active_limits_2_out = Constraint(List_GDTO, rule = genset_power_active_limits_rule_out_2)

def genset_power_reactive_limits_rule_out_1(model,i, t, o):
    return(model_cs.QG_out[i, t, o] >= QG_min[i])
model_cs.genset_power_reactive_limits_1_out = Constraint(List_GDTO, rule = genset_power_reactive_limits_rule_out_1)

def genset_power_reactive_limits_rule_out_2(model,i, t, o):
    return(model_cs.QG_out[i, t, o] <= QG_max[i])
model_cs.genset_power_reactive_limits_2_out = Constraint(List_GDTO, rule = genset_power_reactive_limits_rule_out_2)

#conferir as restries abaixo

#
#def genset_operation_1_rule_out(model,i, t, o):
#    return(model_cs.PGa_out[i, t, o] == 0)
#model_cs.genset_operation_1_out = Constraint(List_GDTO, rule = genset_operation_1_rule_out)
#
#def genset_operation_2_rule_out(model,i, t, o):
#    return(model_cs.PGb_out[i, t, o] == 0)
#model_cs.genset_operation_2_out = Constraint(List_GDTO, rule = genset_operation_2_rule_out)
#
#def genset_operation_3_rule_out(model,i, t, o):
#    return(model_cs.PGc_out[i, t, o] == 0)
#model_cs.genset_operation_3_out = Constraint(List_GDTO, rule = genset_operation_3_rule_out)
#
#def genset_operation_4_rule_out(model,i, t, o):
#    return(model_cs.QGa_out[i, t, o] == 0)
#model_cs.genset_operation_4_out = Constraint(List_GDTO, rule = genset_operation_4_rule_out)
#
#def genset_operation_5_rule_out(model,i, t, o):
#    return(model_cs.QGb_out[i, t, o] == 0)
#model_cs.genset_operation_5_out = Constraint(List_GDTO, rule = genset_operation_5_rule_out)
#
#def genset_operation_6_rule_out(model,i, t, o):
#    return(model_cs.QGc_out[i, t, o] == 0)
#model_cs.genset_operation_6_out = Constraint(List_GDTO, rule = genset_operation_6_rule_out)

# FIX Variables: Islanded operation

model_cs.islanded_operation = ConstraintList()
for i in N:
    for t in T:
        for o in O:
            if int(t) >= int(o) and int(t) < int(o) + 2:
                model_cs.islanded_operation.add(expr = model_cs.Ppcc_a_out[i,t,o] == 0)
                model_cs.islanded_operation.add(expr = model_cs.Ppcc_b_out[i,t,o] == 0)
                model_cs.islanded_operation.add(expr = model_cs.Ppcc_c_out[i,t,o] == 0)
                model_cs.islanded_operation.add(expr = model_cs.Qpcc_a_out[i,t,o] == 0)
                model_cs.islanded_operation.add(expr = model_cs.Qpcc_b_out[i,t,o] == 0)
                model_cs.islanded_operation.add(expr = model_cs.Qpcc_c_out[i,t,o] == 0) 

def fix_voltage_a_rule_out(model_cs,i,t,o):
    return(model_cs.Va_out[i,t,o] == Vnom)
model_cs.fix_voltage_a_out = Constraint(List_NTO, rule = fix_voltage_a_rule_out)

def fix_voltage_b_rule_out(model_cs,i,t,o):
    return(model_cs.Vb_out[i,t,o] == Vnom)
model_cs.fix_voltage_b_out = Constraint(List_NTO, rule = fix_voltage_b_rule_out)

def fix_voltage_c_rule_out(model_cs,i,t,o):
    return(model_cs.Vc_out[i,t,o] == Vnom)
model_cs.fix_voltage_c_out = Constraint(List_NTO, rule = fix_voltage_c_rule_out)

model_cs.fix_active_power_out = ConstraintList()
for i in Tb:
    for t in T:
        for o in O:
            if Tb[i] != 1:
                model_cs.fix_active_power_out.add(expr = model_cs.Ppcc_a_out[i,t,o] == 0)
                model_cs.fix_active_power_out.add(expr = model_cs.Ppcc_b_out[i,t,o] == 0)
                model_cs.fix_active_power_out.add(expr = model_cs.Ppcc_c_out[i,t,o] == 0)
            
model_cs.fix_reactive_power_out = ConstraintList()
for i in Tb:
    for t in T:
        for o in O:
            if Tb[i] != 1:
                model_cs.fix_reactive_power_out.add(expr = model_cs.Qpcc_a_out[i,t,o] == 0)
                model_cs.fix_reactive_power_out.add(expr = model_cs.Qpcc_b_out[i,t,o] == 0)
                model_cs.fix_reactive_power_out.add(expr = model_cs.Qpcc_c_out[i,t,o] == 0)
```

## Constraints **Active Power Losses:**

$

$

```python
def active_losses_a_rule(model, i, j, t):
    expr = (
        (1 / (Va_0[i, t] * Va_0[i, t])) * 
        (Raa_p[i, j] * model.Pa_sqr[i, j, t] + 
         Raa_p[i, j] * model.Qa_sqr[i, j, t] - 
         Xaa_p[i, j] * Pa_0[i, j, t] * model.Qa[i, j, t] + 
         Xaa_p[i, j] * Qa_0[i, j, t] * model.Pa[i, j, t]) +
        (1 / (Va_0[i, t] * Vb_0[i, t])) * 
        (Rab_p[i, j] * Pb_0[i, j, t] * model.Pa[i, j, t] + 
         Rab_p[i, j] * Qb_0[i, j, t] * model.Qa[i, j, t] - 
         Xab_p[i, j] * Pb_0[i, j, t] * model.Qa[i, j, t] + 
         Xab_p[i, j] * Qb_0[i, j, t] * model.Pa[i, j, t]) +
        (1 / (Va_0[i, t] * Vc_0[i, t])) * 
        (Rac_p[i, j] * Pc_0[i, j, t] * model.Pa[i, j, t] + 
         Rac_p[i, j] * Qc_0[i, j, t] * model.Qa[i, j, t] - 
         Xac_p[i, j] * Pc_0[i, j, t] * model.Qa[i, j, t] + 
         Xac_p[i, j] * Qc_0[i, j, t] * model.Pa[i, j, t])
    )
    
    return expr == model.Plss_a[i, j, t]

model.active_losses_a = Constraint(List_LT, rule=active_losses_a_rule)

def active_losses_b_rule(model, i, j, t):
    expr = (
        (1 / (Vb_0[i, t] * Va_0[i, t])) * 
        (Rba_p[i, j] * Pa_0[i, j, t] * model.Pb[i, j, t] + 
         Rba_p[i, j] * Qa_0[i, j, t] * model.Qb[i, j, t] - 
         Xba_p[i, j] * Pa_0[i, j, t] * model.Qb[i, j, t] + 
         Xba_p[i, j] * Qa_0[i, j, t] * model.Pb[i, j, t]) +
        (1 / (Vb_0[i, t] * Vb_0[i, t])) * 
        (Rbb_p[i, j] * model.Pb_sqr[i, j, t]  + 
         Rbb_p[i, j] * model.Qb_sqr[i, j, t]  - 
         Xbb_p[i, j] * Pb_0[i, j, t] * model.Qb[i, j, t] + 
         Xbb_p[i, j] * Qb_0[i, j, t] * model.Pb[i, j, t]) +
        (1 / (Vb_0[i, t] * Vc_0[i, t])) * 
        (Rbc_p[i, j] * Pc_0[i, j, t] * model.Pb[i, j, t] + 
         Rbc_p[i, j] * Qc_0[i, j, t] * model.Qb[i, j, t] - 
         Xbc_p[i, j] * Pc_0[i, j, t] * model.Qb[i, j, t] + 
         Xbc_p[i, j] * Qc_0[i, j, t] * model.Pb[i, j, t])
    )
    
    return expr == model.Plss_b[i, j, t]

model.active_losses_b = Constraint(List_LT, rule=active_losses_b_rule)

def active_losses_c_rule(model, i, j, t):
    expr = (
        (1 / (Vc_0[i, t] * Va_0[i, t])) * 
        (Rca_p[i, j] * Pa_0[i, j, t] * model.Pc[i, j, t] + 
         Rca_p[i, j] * Qa_0[i, j, t] * model.Qc[i, j, t] - 
         Xca_p[i, j] * Pa_0[i, j, t] * model.Qc[i, j, t] + 
         Xca_p[i, j] * Qa_0[i, j, t] * model.Pc[i, j, t]) +
        (1 / (Vc_0[i, t] * Vb_0[i, t])) * 
        (Rcb_p[i, j] * Pb_0[i, j, t] * model.Pc[i, j, t] + 
         Rcb_p[i, j] * Qb_0[i, j, t] * model.Qc[i, j, t] - 
         Xcb_p[i, j] * Pb_0[i, j, t] * model.Qc[i, j, t] + 
         Xcb_p[i, j] * Qb_0[i, j, t] * model.Pc[i, j, t]) +
        (1 / (Vc_0[i, t] * Vc_0[i, t])) * 
        (Rcc_p[i, j] * model.Pc_sqr[i, j, t]  + 
         Rcc_p[i, j] * model.Qc_sqr[i, j, t]  - 
         Xcc_p[i, j] * Pc_0[i, j, t] * model.Qc[i, j, t] + 
         Xcc_p[i, j] * Qc_0[i, j, t] * model.Pc[i, j, t])
    )
    
    return expr == model.Plss_c[i, j, t]

model.active_losses_c = Constraint(List_LT, rule=active_losses_c_rule)
```

\sum_{ki\in \mathcal{L}} P_{ki,f,t} - \sum_{ij\in \mathcal{L}} (P_{ij,f,t} + P^{\text{L}}_{ij,f,t}) + P^\text{S}_{i,f,t} + \sum_{n\in \mathcal{G} |i=L^G(n)} P^{\text{G}}_{n,f,t} = P^\text{D}_{i,f,t} - P^\text{PV}_{i,f,t} + \sum_{m\in \mathcal{B} |i=L^B(m)} P^{\text{B}}_{m,t} + \sum_{r\in \mathcal{V} |i=L^{\text{EV}}(r)} P^{\text{\text{EV}}}_{r,f,t} \ \ \forall ij,f,t
$



```python
def active_power_balance_rule_a(model, i,t):
    return (
        sum(model.Pa[a, j, t]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Plss_a[a, j, t]*p[(i,a,j)] for (a,j) in L) +
        model.Ppcc_a[i, t] +
        sum(model.PGa[a, t] for a in dict_nos_gd[i]) +
        sum(model.PB_dis_a[a, t] for a in dict_nos_bs[i]) -
        sum(model.PB_ch_a[a, t] for a in dict_nos_bs[i]) -
        sum(model.PEV_ch_a[a, t] for a in dict_nos_ev[i]) -
        PDa[(i,t)] + PVa[(i,t)] == 0
    )

model.active_power_balance_a = Constraint(List_NT, rule=active_power_balance_rule_a)

def active_power_balance_rule_b(model, i,t):
    return (
        sum(model.Pb[a, j, t]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Plss_b[a, j, t]*p[(i,a,j)] for (a,j) in L) +
        model.Ppcc_b[i, t] +
        sum(model.PGb[a, t] for a in dict_nos_gd[i]) +
        sum(model.PB_dis_b[a, t] for a in dict_nos_bs[i]) -
        sum(model.PB_ch_b[a, t] for a in dict_nos_bs[i]) -
        sum(model.PEV_ch_b[a, t] for a in dict_nos_ev[i]) -
        PDb[(i,t)] + PVb[(i,t)] == 0
    )

model.active_power_balance_b = Constraint(List_NT, rule=active_power_balance_rule_b)

def active_power_balance_rule_c(model, i,t):
    return (
        sum(model.Pc[a, j, t]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Plss_c[a, j, t]*p[(i,a,j)] for (a,j) in L) +
        model.Ppcc_c[i, t] +
        sum(model.PGc[a, t] for a in dict_nos_gd[i]) +
        sum(model.PB_dis_c[a, t] for a in dict_nos_bs[i]) -
        sum(model.PB_ch_c[a, t] for a in dict_nos_bs[i]) -
        sum(model.PEV_ch_c[a, t] for a in dict_nos_ev[i]) -
        PDc[(i,t)] + PVc[(i,t)] == 0
    )

model.active_power_balance_c = Constraint(List_NT, rule=active_power_balance_rule_c)
```

## Constraint **Reactive Power Balance**:

$
\sum_{ki\in \mathcal{L}} Q_{ki,f,t} - \sum_{ij\in \mathcal{L}} (Q_{ij,f,t} + Q^{\text{L}}_{ij,f,t}) + Q^\text{S}_{i,f,t} + \sum_{n\in \mathcal{G} |i=L^G(n)} Q^{\text{G}}_{n,f,t}= Q^\text{D}_{i,f,t} \ \ \forall ij,f,t
$




```python
def reactive_power_balance_rule_a(model, i,t):
    return (
        sum(model.Qa[a, j, t]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Qlss_a[a, j, t]*p[(i,a,j)] for (a,j) in L) +
        model.Qpcc_a[i, t] +
        sum(model.QGa[a, t] for a in dict_nos_gd[i]) -
        QDa[(i,t)] == 0
    )

model.reactive_power_balance_a = Constraint(List_NT, rule=reactive_power_balance_rule_a)

def reactive_power_balance_rule_b(model, i,t):
    return (
        sum(model.Qb[a, j, t]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Qlss_b[a, j, t]*p[(i,a,j)] for (a,j) in L) +
        model.Qpcc_b[i, t] +
        sum(model.QGb[a, t] for a in dict_nos_gd[i]) -
        QDb[(i,t)] == 0
    )

model.reactive_power_balance_b = Constraint(List_NT, rule=reactive_power_balance_rule_b)

def reactive_power_balance_rule_c(model, i,t):
    return (
        sum(model.Qc[a, j, t]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Qlss_c[a, j, t]*p[(i,a,j)] for (a,j) in L) +
        model.Qpcc_c[i, t] +
        sum(model.QGc[a, t] for a in dict_nos_gd[i]) -
        QDc[(i,t)] == 0
    )

model.reactive_power_balance_c = Constraint(List_NT, rule=reactive_power_balance_rule_c)
```

## Constraint **Voltage Drop in the lines**:

$
V^{\text{2}}_{i,f,t} - V^{\text{2}}_{j,f,t} = 2 \sum_{h \in \mathcal{F}} \left(  R'_{ij,f,h} P_{ij,h,t} + X'_{ij,f,h} Q_{ij,h,t} \right) -  \frac{1}{V^{\text{2}}_{i,f,t}} \left[\sum_{h \in \mathcal{F}}\left(R'_{ij,f,h}P_{ij,h,t} + X'_{ij,f,h}Q_{ij,h,t}\right) \right]^2 -\frac{1}{V^{\text{2}}_{i,f,t}}\left[\sum_{h \in \mathcal{F}}\left(R'_{ij,f,h}Q_{ij,h,t} - X'_{ij,f,h}P_{ij,h,t}\right) \right]^2  \forall ij,f,h,t
$


```python
def voltage_droop_rule_a(model, i,j,t):
    return((2*(Raa_p[i,j]*model.Pa[i,j,t] + Xaa_p[i,j]*model.Qa[i,j,t])) + 
           (2*(Rab_p[i,j]*model.Pb[i,j,t] + Xab_p[i,j]*model.Qb[i,j,t])) + 
           (2*(Rac_p[i,j]*model.Pc[i,j,t] + Xac_p[i,j]*model.Qc[i,j,t])) -
           (1/Va_0[i,t]**2) * ((Raa_p[i,j]**2 + Xaa_p[i,j]**2) * (model.Pa_sqr[i,j,t] + model.Qa_sqr[i,j,t])) ==
           model.Va_sqr[i,t] - model.Va_sqr[j,t])
model.voltage_droop_a = Constraint(List_LT, rule= voltage_droop_rule_a)

def voltage_droop_rule_b(model, i,j,t):
    return((2*(Rba_p[i,j]*model.Pa[i,j,t] + Xba_p[i,j]*model.Qa[i,j,t])) + 
           (2*(Rbb_p[i,j]*model.Pb[i,j,t] + Xbb_p[i,j]*model.Qb[i,j,t])) + 
           (2*(Rbc_p[i,j]*model.Pc[i,j,t] + Xbc_p[i,j]*model.Qc[i,j,t])) -
           (1/Vb_0[i,t]**2) * ((Rbb_p[i,j]**2 + Xbb_p[i,j]**2) * (model.Pb_sqr[i,j,t] + model.Qb_sqr[i,j,t])) ==
           model.Vb_sqr[i,t] - model.Vb_sqr[j,t])
model.voltage_droop_b = Constraint(List_LT, rule= voltage_droop_rule_b)

def voltage_droop_rule_c(model, i,j,t):
    return((2*(Rca_p[i,j]*model.Pa[i,j,t] + Xca_p[i,j]*model.Qa[i,j,t])) + 
           (2*(Rcb_p[i,j]*model.Pb[i,j,t] + Xcb_p[i,j]*model.Qb[i,j,t])) + 
           (2*(Rcc_p[i,j]*model.Pc[i,j,t] + Xcc_p[i,j]*model.Qc[i,j,t])) -
           (1/Vc_0[i,t]**2) * ((Rcc_p[i,j]**2 + Xcc_p[i,j]**2) * (model.Pc_sqr[i,j,t] + model.Qc_sqr[i,j,t])) ==
           model.Vc_sqr[i,t] - model.Vc_sqr[j,t])
model.voltage_droop_c = Constraint(List_LT, rule= voltage_droop_rule_c)
```

## Constraint **Current Limits**:

\frac{(P^{2}_{ij,f,t} + Q^{2}_{ij,f,t})}{V^{2}_{i,f,t}} \leq \overline{I}_{ij}^2 \ \ \forall ij,f,t
$

```python
def current_limits_rule_a(model, i,j,t):
    return((model.Pa_sqr[i,j,t] + model.Qa_sqr[i,j,t]) <= Imax[i,j]**2 * (model.Va_sqr[i,t]))
model.current_limits_a = Constraint(List_LT, rule = current_limits_rule_a)

def current_limits_rule_b(model, i,j,t):
    return((model.Pb_sqr[i,j,t] + model.Qb_sqr[i,j,t]) <= Imax[i,j]**2 * (model.Vb_sqr[i,t]))
model.current_limits_b = Constraint(List_LT, rule = current_limits_rule_b)

def current_limits_rule_c(model, i,j,t):
    return((model.Pc_sqr[i,j,t] + model.Qc_sqr[i,j,t]) <= Imax[i,j]**2 * (model.Vc_sqr[i,t]))
model.current_limits_c = Constraint(List_LT, rule = current_limits_rule_c)
```

## Constraint **Piecewise linearization of power flow:**

```python
# Active power linearization constraints

def pa_calculation_rule(model, i,j,t):
    return( model.Pa_sqr[i,j,t] - sum(S_ms[i,j,y] * model.Pa_Dp[i,j,t,y] for y in Y) == 0)
model.Pa_calculation = Constraint(List_LT, rule = pa_calculation_rule)

def pb_calculation_rule(model, i,j,t):
    return( model.Pb_sqr[i,j,t] - sum(S_ms[i,j,y] * model.Pb_Dp[i,j,t,y] for y in Y) == 0)
model.Pb_calculation = Constraint(List_LT, rule = pb_calculation_rule)

def pc_calculation_rule(model, i,j,t):
    return( model.Pc_sqr[i,j,t] - sum(S_ms[i,j,y] * model.Pc_Dp[i,j,t,y] for y in Y) == 0)
model.Pc_calculation = Constraint(List_LT, rule = pc_calculation_rule)

# Reactive power linearization constraints

def qa_calculation_rule(model, i,j,t):
    return( model.Qa_sqr[i,j,t] - sum(S_ms[i,j,y] * model.Qa_Dp[i,j,t,y] for y in Y) == 0)
model.Qa_calculation = Constraint(List_LT, rule = qa_calculation_rule)

def qb_calculation_rule(model, i,j,t):
    return( model.Qb_sqr[i,j,t] - sum(S_ms[i,j,y] * model.Qb_Dp[i,j,t,y] for y in Y) == 0)
model.Qb_calculation = Constraint(List_LT, rule = qb_calculation_rule)

def qc_calculation_rule(model, i,j,t):
    return( model.Qc_sqr[i,j,t] - sum(S_ms[i,j,y] * model.Qc_Dp[i,j,t,y] for y in Y) == 0)
model.Qc_calculation = Constraint(List_LT, rule = qc_calculation_rule)


# Linearizations variables

def pa_p_rule(model, i,j,t):
    return( model.Pa_p[i,j,t] - model.Pa_n[i,j,t] - model.Pa[i,j,t] == 0)
model.Pa_p_constraint = Constraint(List_LT, rule = pa_p_rule)

def pb_p_rule(model, i,j,t):
    return( model.Pb_p[i,j,t] - model.Pb_n[i,j,t] - model.Pb[i,j,t] == 0)
model.Pb_p_constraint = Constraint(List_LT, rule = pb_p_rule)

def pc_p_rule(model, i,j,t):
    return( model.Pc_p[i,j,t] - model.Pc_n[i,j,t] - model.Pc[i,j,t] == 0)
model.Pc_p_constraint = Constraint(List_LT, rule = pc_p_rule)

def qa_p_rule(model, i,j,t):
    return( model.Qa_p[i,j,t] - model.Qa_n[i,j,t] - model.Qa[i,j,t] == 0)
model.Qa_p_constraint = Constraint(List_LT, rule = qa_p_rule)

def qb_p_rule(model, i,j,t):
    return( model.Qb_p[i,j,t] - model.Qb_n[i,j,t] - model.Qb[i,j,t] == 0)
model.Qb_p_constraint = Constraint(List_LT, rule = qb_p_rule)

def qc_p_rule(model, i,j,t):
    return( model.Qc_p[i,j,t] - model.Qc_n[i,j,t] - model.Qc[i,j,t] == 0)
model.Qc_p_constraint = Constraint(List_LT, rule = qc_p_rule)

def pa_abs_rule(model, i,j,t):
    return( model.Pa_p[i,j,t] + model.Pa_n[i,j,t] - sum(model.Pa_Dp[i,j,t,y] for y in Y) == 0)
model.pa_abs = Constraint(List_LT, rule = pa_abs_rule)

def pb_abs_rule(model, i,j,t):
    return( model.Pb_p[i,j,t] + model.Pb_n[i,j,t] - sum(model.Pb_Dp[i,j,t,y] for y in Y) == 0)
model.pb_abs = Constraint(List_LT, rule = pb_abs_rule)

def pc_abs_rule(model, i,j,t):
    return( model.Pc_p[i,j,t] + model.Pc_n[i,j,t] - sum(model.Pc_Dp[i,j,t,y] for y in Y) == 0)
model.pc_abs = Constraint(List_LT, rule = pc_abs_rule)

def qa_abs_rule(model, i,j,t):
    return( model.Qa_p[i,j,t] + model.Qa_n[i,j,t] - sum(model.Qa_Dp[i,j,t,y] for y in Y) == 0)
model.qa_abs = Constraint(List_LT, rule = qa_abs_rule)

def qb_abs_rule(model, i,j,t):
    return( model.Qb_p[i,j,t] + model.Qb_n[i,j,t] - sum(model.Qb_Dp[i,j,t,y] for y in Y) == 0)
model.qb_abs = Constraint(List_LT, rule = qb_abs_rule)

def qc_abs_rule(model, i,j,t):
    return( model.Qc_p[i,j,t] + model.Qc_n[i,j,t] - sum(model.Qc_Dp[i,j,t,y] for y in Y) == 0)
model.qc_abs = Constraint(List_LT, rule = qc_abs_rule)

def pa_limits_rule(model, i,j,t,y):
    return( model.Pa_Dp[i,j,t,y] <= S_Dp_max[i,j])
model.pa_limits = Constraint(List_LTY, rule = pa_limits_rule)

def pb_limits_rule(model, i,j,t,y):
    return( model.Pb_Dp[i,j,t,y] <= S_Dp_max[i,j])
model.pb_limits = Constraint(List_LTY, rule = pb_limits_rule)

def pc_limits_rule(model, i,j,t,y):
    return( model.Pc_Dp[i,j,t,y] <= S_Dp_max[i,j])
model.pc_limits = Constraint(List_LTY, rule = pc_limits_rule)

def qa_limits_rule(model, i,j,t,y):
    return( model.Qa_Dp[i,j,t,y] <= S_Dp_max[i,j])
model.qa_limits = Constraint(List_LTY, rule = qa_limits_rule)

def qb_limits_rule(model, i,j,t,y):
    return( model.Qb_Dp[i,j,t,y] <= S_Dp_max[i,j])
model.qb_limits = Constraint(List_LTY, rule = qb_limits_rule)

def qc_limits_rule(model, i,j,t,y):
    return( model.Qc_Dp[i,j,t,y] <= S_Dp_max[i,j])
model.qc_limits = Constraint(List_LTY, rule = qc_limits_rule)
```

## Constraint **Genset:**

$
P^{\text{G}}_{n,t} =  \sum_{f \in \mathcal{F}} P^{\text{G}}_{n,f,t}
\;\;\;\; \forall n,f,t $

$
Q^{\text{G}}_{n,t} =  \sum_{f \in \mathcal{F}} Q^{\text{G}}_{n,f,t}
\;\;\;\; \forall n,f,t$

$
\underline{P}^{\text{G}}_n \cdot u_{n,t} \leq P^{\text{G}}_{n,t} \leq \overline{P}^{\text{G}}_n  \cdot u_{n,t}
\;\;\;\; \forall n,f,t$

$
\underline{Q}^{\text{G}}_n \cdot u_{n,t} \leq Q^{\text{G}}_{n,t} \leq \overline{Q}^{\text{G}}_n  \cdot u_{n,t}
\;\;\;\; \forall n,f,t$

```python
def genset_power_rule(model,i,t):
    return(model.PGa[i,t] + model.PGb[i,t] + model.PGc[i,t] == model.PG[i,t])
model.genset_power_active = Constraint(List_GDT, rule=genset_power_rule)

def genset_power_reactive_rule(model,i,t):
    return(model.QGa[i,t] + model.QGb[i,t] + model.QGc[i,t] == model.QG[i,t])
model.genset_power_reactive = Constraint(List_GDT, rule=genset_power_reactive_rule)

def genset_power_active_limits_rule_1(model,i,t):
    return(model.PG[i,t] >= PG_min[i]*model.oG[i,t])
model.genset_power_active_limits_1 = Constraint(List_GDT, rule = genset_power_active_limits_rule_1)

def genset_power_active_limits_rule_2(model,i,t):
    return(model.PG[i,t] <= PG_max[i]*model.oG[i,t])
model.genset_power_active_limits_2 = Constraint(List_GDT, rule = genset_power_active_limits_rule_2)

def genset_power_reactive_limits_rule_1(model,i,t):
    return(model.QG[i,t] >= QG_min[i]*model.oG[i,t])
model.genset_power_reactive_limits_1 = Constraint(List_GDT, rule = genset_power_reactive_limits_rule_1)

def genset_power_reactive_limits_rule_2(model,i,t):
    return(model.QG[i,t] <= QG_max[i]*model.oG[i,t])
model.genset_power_reactive_limits_2 = Constraint(List_GDT, rule = genset_power_reactive_limits_rule_2)
```

## Constraint **BESS:**

$
E_{m,t}^\text{B}  = E_{m}^\text{B,\text{ini}}+ P^\text{B}_{m,t} \cdot \Delta t \;\;\;\; \forall m,t | _{t = 1}
$

$
E_{m,t}^\text{B}  = E_{m,t-1}^\text{B}+ P^\text{B}_{m,t} \cdot \Delta t
\;\;\;\; \forall m, t | _{t>1} $

$
P_{m,t}^\text{B}  = P^\text{B+}_{m,t} \cdot \eta_{m}^\text{B} -P_{m,t}^\text{B-} \cdot \frac{1}{\eta_{m}^\text{B}}
\;\;\;\; \forall m, t$

$
P^\text{B+}_{m,t} \cdot P_{m,t}^\text{B-} = 0
\;\;\;\; \forall m, t$

$
P^{\text{B-}}_{m,t} =  \sum_{f \in \mathcal{F}} P^{\text{B-}}_{m,f,t}
\;\;\;\; \forall m, t$

$
P^{\text{B+}}_{m,t} =  \sum_{f \in \mathcal{F}} P^{\text{B+}}_{m,f,t}
\;\;\;\; \forall m, t$

$
P^{\text{B+}}_{m,f,t} =  P^{\text{B+}}_{m,h,t}
\;\;\;\;\forall m, f, h, t$

$
P^{\text{B-}}_{m,f,t} =  P^{\text{B-}}_{m,h,t}
\;\;\;\;\forall m, f, h, t$

$
0 \leq P^{\text{B+}}_{m,t} \leq \overline{P}^{\text{B}}_{m}
\;\;\;\; \forall m, t$

$
0 \leq P^{\text{B-}}_{m,t} \leq \overline{P}^{\text{B}}_{m}
\;\;\;\; \forall m, t$

$
\underline{E}^{\text{B}}_{m} \leq E^{\text{B}}_{m,t} \leq \overline{E}^{\text{B}}_{m}
\;\;\;\; \forall m, t$

## Constraint **EV:**

E_{r,t}^\text{EV}  = E_{r,t-1}^\text{EV}+ P^\text{EV}_{r,t} \cdot \eta_{r}^\text{EV} \cdot \Delta t
\;\;\;\; \forall r, t  | _{ t > t_{a} \ \text{and} \ t \leq t_d}$

$
P^{\text{EV+}}_{r,t} =  \sum_{f \in \mathcal{F}} P^{\text{EV+}}_{r,f,t}
\;\;\;\; \forall r, t  | _{ t \geq t_{a} \ \text{and} \ t \leq t_d}$

$
P^{\text{EV+}}_{r,f,t} =  P^{\text{EV+}}_{r,h,t}
\;\;\;\; \forall r, t  | _{ t \geq t_{a} \ \text{and} \ t \leq t_d}$

0 \leq P^{\text{EV+}}_{r,t} \leq \overline{P}^{\text{EV}}_{r}

\;\;\;\; \forall r, t  | _{ t \geq t_{a} \ \text{and} \ t \leq t_d}$

$
E^{\text{EV}}_{r,t} = \overline{E}^{\text{EV}}_{r}
\;\;\;\; \forall r,t | _{t = t_d}$

$
P^{\text{EV+}}_{r,f,t} = 0 
\;\;\;\;\forall r,f,t | _{ t < t_{a} \ \text{or} \ t > t_d}$

## Constraints **Active Power Balance**:

$
\sum_{ki\in \mathcal{L}} P_{ki,f,t} - \sum_{ij\in \mathcal{L}} (P_{ij,f,t} + P^{\text{L}}_{ij,f,t}) + P^\text{S}_{i,f,t} + \sum_{n\in \mathcal{G} |i=L^G(n)} P^{\text{G}}_{n,f,t} = P^\text{D}_{i,f,t} - P^\text{PV}_{i,f,t} + \sum_{m\in \mathcal{B} |i=L^B(m)} P^{\text{B}}_{m,t} + \sum_{r\in \mathcal{V} |i=L^{\text{EV}}(r)} P^{\text{\text{EV}}}_{r,f,t} \ \ \forall ij,f,t
$



```python
def active_power_balance_rule_out_a(model, i, t, o):
    return (
        sum(model.Pa_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) -  
        sum(model.Plss_a_out[a, j, t, o]*p[(i,a,j)] for (a,j) in L) +
        model.Ppcc_a_out[i, t, o] +
        sum(model.PGa_out[a, t, o] for a in dict_nos_gd[i]) +
        sum(model.PB_dis_a[a, t] for a in dict_nos_bs[i]) -
        sum(model.PB_ch_a[a, t] for a in dict_nos_bs[i]) -
        sum(model.PEV_ch_a[a, t] for a in dict_nos_ev[i]) +
        (- PDa[(i,t)] + PVa[(i,t)]) * model.xd[i, t, o] == 0
    )

model.active_power_balance_a_out = Constraint(List_NTO, rule=active_power_balance_rule_out_a)

def active_power_balance_rule_out_b(model, i,t,o):
    return (
        sum(model.Pb_out[a, j, t,o]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Plss_b_out[a, j, t,o]*p[(i,a,j)] for (a,j) in L) +
        model.Ppcc_b_out[i, t,o] +
        sum(model.PGb_out[a, t, o] for a in dict_nos_gd[i]) +
        sum(model.PB_dis_b[a, t] for a in dict_nos_bs[i]) -
        sum(model.PB_ch_b[a, t] for a in dict_nos_bs[i]) -
        sum(model.PEV_ch_b[a, t] for a in dict_nos_ev[i]) +
        (- PDb[(i,t)] + PVb[(i,t)]) * model.xd[i, t, o] == 0
    )

model.active_power_balance_b_out = Constraint(List_NTO, rule=active_power_balance_rule_out_b)

def active_power_balance_rule_out_c(model, i,t,o):
    return (
        sum(model.Pc_out[a, j, t,o]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Plss_c_out[a, j, t,o]*p[(i,a,j)] for (a,j) in L) +
        model.Ppcc_c_out[i, t,o] +
        sum(model.PGc_out[a, t,o] for a in dict_nos_gd[i]) +
        sum(model.PB_dis_c[a, t] for a in dict_nos_bs[i]) -
        sum(model.PB_ch_c[a, t] for a in dict_nos_bs[i]) -
        sum(model.PEV_ch_c[a, t] for a in dict_nos_ev[i]) +
        (- PDc[(i,t)] + PVc[(i,t)]) * model.xd[i, t, o] == 0
    )

model.active_power_balance_c_out = Constraint(List_NTO, rule=active_power_balance_rule_out_c)   
```

## Constraint **Reactive Power Balance**:

$
\sum_{ki\in \mathcal{L}} Q_{ki,f,t} - \sum_{ij\in \mathcal{L}} (Q_{ij,f,t} + Q^{\text{L}}_{ij,f,t}) + Q^\text{S}_{i,f,t} + \sum_{n\in \mathcal{G} |i=L^G(n)} Q^{\text{G}}_{n,f,t}= Q^\text{D}_{i,f,t} \ \ \forall ij,f,t
$




```python
def reactive_power_balance_out_rule_out_a(model, i,t,o):
    return (
        sum(model.Qa_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Qlss_a_out[a, j, t, o]*p[(i,a,j)] for (a,j) in L) +
        model.Qpcc_a_out[i, t,o] +
        sum(model.QGa_out[a, t,o] for a in dict_nos_gd[i]) -
        QDa[(i,t)] * model.xd[i, t, o] == 0
    )

model.reactive_power_balance_out_a = Constraint(List_NTO, rule=reactive_power_balance_out_rule_out_a)

def reactive_power_balance_out_rule_out_b(model, i,t,o):
    return (
        sum(model.Qb_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Qlss_b_out[a, j, t, o]*p[(i,a,j)] for (a,j) in L) +
        model.Qpcc_b_out[i, t,o] +
        sum(model.QGb_out[a, t,o] for a in dict_nos_gd[i]) -
        QDb[(i,t)] * model.xd[i, t, o] == 0
    )

model.reactive_power_balance_out_b = Constraint(List_NTO, rule=reactive_power_balance_out_rule_out_b)

def reactive_power_balance_out_rule_out_c(model, i,t,o):
    return (
        sum(model.Qc_out[a, j, t, o]*df[(i,a,j)] for (a,j) in L) -
        sum(model.Qlss_c_out[a, j, t, o]*p[(i,a,j)] for (a,j) in L) +
        model.Qpcc_c_out[i, t,o] +
        sum(model.QGc_out[a, t,o] for a in dict_nos_gd[i]) -
        QDc[(i,t)] * model.xd[i, t, o] == 0
    )

model.reactive_power_balance_out_c = Constraint(List_NTO, rule=reactive_power_balance_out_rule_out_c)

```

## Constraint **Voltage Droop in the lines**:

$
V^{\text{2}}_{i,f,t} - V^{\text{2}}_{j,f,t} = 2 \sum_{h \in \mathcal{F}} \left(  R'_{ij,f,h} P_{ij,h,t} + X'_{ij,f,h} Q_{ij,h,t} \right) -  \frac{1}{V^{\text{2}}_{i,f,t}} \left[\sum_{h \in \mathcal{F}}\left(R'_{ij,f,h}P_{ij,h,t} + X'_{ij,f,h}Q_{ij,h,t}\right) \right]^2 -\frac{1}{V^{\text{2}}_{i,f,t}}\left[\sum_{h \in \mathcal{F}}\left(R'_{ij,f,h}Q_{ij,h,t} - X'_{ij,f,h}P_{ij,h,t}\right) \right]^2  \forall ij,f,h,t
$


```python
def voltage_droop_out_rule_out_a(model, i,j,t,o):
    return((2*(Raa_p[i,j]*model.Pa_out[i,j,t,o] + Xaa_p[i,j]*model.Qa_out[i,j,t,o])) + 
           (2*(Rab_p[i,j]*model.Pb_out[i,j,t,o] + Xab_p[i,j]*model.Qb_out[i,j,t,o])) + 
           (2*(Rac_p[i,j]*model.Pc_out[i,j,t,o] + Xac_p[i,j]*model.Qc_out[i,j,t,o])) -
           (1/Va_0_out[i,t,o]**2) * ((Raa_p[i,j]**2 + Xaa_p[i,j]**2) * (model.Pa_sqr_out[i,j,t,o] + model.Qa_sqr_out[i,j,t,o])) ==
           model.Va_sqr_out[i,t,o] - model.Va_sqr_out[j,t,o])
model.voltage_droop_out_a = Constraint(List_LTO, rule= voltage_droop_out_rule_out_a)

def voltage_droop_out_rule_out_b(model, i,j,t,o):
    return((2*(Rba_p[i,j]*model.Pa_out[i,j,t,o] + Xba_p[i,j]*model.Qa_out[i,j,t,o])) + 
           (2*(Rbb_p[i,j]*model.Pb_out[i,j,t,o] + Xbb_p[i,j]*model.Qb_out[i,j,t,o])) + 
           (2*(Rbc_p[i,j]*model.Pc_out[i,j,t,o] + Xbc_p[i,j]*model.Qc_out[i,j,t,o])) -
           (1/Vb_0_out[i,t,o]**2) * ((Rbb_p[i,j]**2 + Xbb_p[i,j]**2) * (model.Pb_sqr_out[i,j,t,o] + model.Qb_sqr_out[i,j,t,o])) ==
           model.Vb_sqr_out[i,t,o] - model.Vb_sqr_out[j,t,o])
model.voltage_droop_out_b = Constraint(List_LTO, rule= voltage_droop_out_rule_out_b)

def voltage_droop_out_rule_out_c(model, i,j,t,o):
    return((2*(Rca_p[i,j]*model.Pa_out[i,j,t,o] + Xca_p[i,j]*model.Qa_out[i,j,t,o])) + 
           (2*(Rcb_p[i,j]*model.Pb_out[i,j,t,o] + Xcb_p[i,j]*model.Qb_out[i,j,t,o])) + 
           (2*(Rcc_p[i,j]*model.Pc_out[i,j,t,o] + Xcc_p[i,j]*model.Qc_out[i,j,t,o])) -
           (1/Vc_0_out[i,t,o]**2) * ((Rcc_p[i,j]**2 + Xcc_p[i,j]**2) * (model.Pc_sqr_out[i,j,t,o] + model.Qc_sqr_out[i,j,t,o])) ==
           model.Vc_sqr_out[i,t,o] - model.Vc_sqr_out[j,t,o])
model.voltage_droop_out_c = Constraint(List_LTO, rule= voltage_droop_out_rule_out_c)
```

## Constraint **Current Limits**:

$

## Constraint **Genset:**

$
P^{\text{G}}_{n,t} =  \sum_{f \in \mathcal{F}} P^{\text{G}}_{n,f,t}
\;\;\;\; \forall n,f,t $

$
Q^{\text{G}}_{n,t} =  \sum_{f \in \mathcal{F}} Q^{\text{G}}_{n,f,t}
\;\;\;\; \forall n,f,t$

$
\underline{P}^{\text{G}}_n \cdot u_{n,t} \leq P^{\text{G}}_{n,t} \leq \overline{P}^{\text{G}}_n  \cdot u_{n,t}
\;\;\;\; \forall n,f,t$

$
\underline{Q}^{\text{G}}_n \cdot u_{n,t} \leq Q^{\text{G}}_{n,t} \leq \overline{Q}^{\text{G}}_n  \cdot u_{n,t}
\;\;\;\; \forall n,f,t$

```python
def genset_power_rule_out(model,i,t,o):
    return(model.PGa_out[i,t,o] + model.PGb_out[i,t,o] + model.PGc_out[i,t,o] == model.PG_out[i,t,o])
model.genset_power_active_out = Constraint(List_GDTO, rule=genset_power_rule_out)

def genset_power_reactive_rule_out(model,i,t,o):
    return(model.QGa_out[i,t,o] + model.QGb_out[i,t,o] + model.QGc_out[i,t,o] == model.QG_out[i,t,o])
model.genset_power_reactive_out = Constraint(List_GDTO, rule=genset_power_reactive_rule_out)

def genset_power_active_limits_rule_out_1(model,i,t,o):
    return(model.PG_out[i,t,o] >= PG_min[i])
model.genset_power_active_limits_1_out = Constraint(List_GDTO, rule = genset_power_active_limits_rule_out_1)

def genset_power_active_limits_rule_out_2(model,i,t,o):
    return(model.PG_out[i,t,o] <= PG_max[i])
model.genset_power_active_limits_2_out = Constraint(List_GDTO, rule = genset_power_active_limits_rule_out_2)

def genset_power_reactive_limits_rule_out_1(model,i,t,o):
    return(model.QG_out[i,t,o] >= QG_min[i])
model.genset_power_reactive_limits_1_out = Constraint(List_GDTO, rule = genset_power_reactive_limits_rule_out_1)

def genset_power_reactive_limits_rule_out_2(model,i,t,o):
    return(model.QG_out[i,t,o] <= QG_max[i])
model.genset_power_reactive_limits_2_out = Constraint(List_GDTO, rule = genset_power_reactive_limits_rule_out_2)

model.genset_operation_grid_connected = ConstraintList()
for i in GD:
    for t in T:
        for o in O:
            if int(t) < int(c) or int(t) >= int(c) + 2:
                model.genset_operation_grid_connected.add(expr = model.PGa_out[i,t,o] == 0)
                model.genset_operation_grid_connected.add(expr = model.PGb_out[i,t,o] == 0)
                model.genset_operation_grid_connected.add(expr = model.PGc_out[i,t,o] == 0)
                model.genset_operation_grid_connected.add(expr = model.QGa_out[i,t,o] == 0)
                model.genset_operation_grid_connected.add(expr = model.QGb_out[i,t,o] == 0)
                model.genset_operation_grid_connected.add(expr = model.QGc_out[i,t,o] == 0)
```

## Constraint Fix Islanded Operation

```python
model.islanded_operation = ConstraintList()
for i in N:
    for t in T:
        for o in O:
            if int(t) >= int(o) and int(t) < int(o) + 2:
                model.islanded_operation.add(expr = model.Ppcc_a_out[i,t,o] == 0)
                model.islanded_operation.add(expr = model.Ppcc_b_out[i,t,o] == 0)
                model.islanded_operation.add(expr = model.Ppcc_c_out[i,t,o] == 0)
                model.islanded_operation.add(expr = model.Qpcc_a_out[i,t,o] == 0)
                model.islanded_operation.add(expr = model.Qpcc_b_out[i,t,o] == 0)
                model.islanded_operation.add(expr = model.Qpcc_c_out[i,t,o] == 0) 
```

## Constraint Fix Variables

```python

model.fix_voltage_out_1 = ConstraintList()
for i in Tb:
    for t in T:
        for o in O:
            if Tb[i] == 1 and int(t) < int(c) or int(t) >= int(c) + 2:
                model.fix_voltage_out_1.add(expr = model.Va_out[i,t,o] == Vnom)
                model.fix_voltage_out_1.add(expr = model.Vb_out[i,t,o] == Vnom)
                model.fix_voltage_out_1.add(expr = model.Vc_out[i,t,o] == Vnom)
                model.fix_voltage_out_1.add(expr = model.Va_sqr_out[i,t,o] == Vnom**2)
                model.fix_voltage_out_1.add(expr = model.Vb_sqr_out[i,t,o] == Vnom**2)
                model.fix_voltage_out_1.add(expr = model.Vc_sqr_out[i,t,o] == Vnom**2)


model.fix_voltage_out_2 = ConstraintList()
for i in Tb:
    for t in T:
        for o in O:
            if Tb[i] == 2 and int(t) >= int(c) and int(t) < int(c) + 2:  
                model.fix_voltage_out_2.add(expr = model.Va_out[i,t,o] == Vnom)
                model.fix_voltage_out_2.add(expr = model.Vb_out[i,t,o] == Vnom)
                model.fix_voltage_out_2.add(expr = model.Vc_out[i,t,o] == Vnom)
                model.fix_voltage_out_2.add(expr = model.Va_sqr_out[i,t,o] == Vnom**2)
                model.fix_voltage_out_2.add(expr = model.Vb_sqr_out[i,t,o] == Vnom**2)
                model.fix_voltage_out_2.add(expr = model.Vc_sqr_out[i,t,o] == Vnom**2)


model.fix_active_power_out = ConstraintList()
for i in Tb:
    for t in T:
        for o in O:
            if Tb[i] != 1:
                model.fix_active_power_out.add(expr = model.Ppcc_a_out[i,t,o] == 0)
                model.fix_active_power_out.add(expr = model.Ppcc_b_out[i,t,o] == 0)
                model.fix_active_power_out.add(expr = model.Ppcc_c_out[i,t,o] == 0)
            
model.fix_reactive_power_out = ConstraintList()
for i in Tb:
    for t in T:
        for o in O:
            if Tb[i] != 1:
                model.fix_reactive_power_out.add(expr = model.Qpcc_a_out[i,t,o] == 0)
                model.fix_reactive_power_out.add(expr = model.Qpcc_b_out[i,t,o] == 0)
                model.fix_reactive_power_out.add(expr = model.Qpcc_c_out[i,t,o] == 0)
```
